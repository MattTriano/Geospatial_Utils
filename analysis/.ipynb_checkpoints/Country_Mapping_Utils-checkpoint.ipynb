## Choropleths: Mapping Data Features

A Choropleth map is a thematic map where each area in the map is colored or patterned to show some property of the area, such as the population, or the continent, or the voting numbers for an election. To communicate this information correctly, you have to choose the **colormap** that is appropriate for the property you want to visualize. The three types of properties/colormaps that you'll use in choropleths are: 

* **Sequential**: This type of colormap is for numerical properties that span from 0 (or some positive minimum value) to some maximum value. Examples of the data that should use a Sequential colormap are counts (eg population, crime, GDP, etc), rates (GDP per capita, crimes per 100k pop, etc), or ordered categoricals (eg rankings, maturity of economy, percent of population in an age-group).
* **Qualitative** (aka **Unordered Categorical**): This type of colormap is for data that is broken down by categories and there isn't an objective ordering to the categories (eg continents, gender, etc).
* **Divergent**: This type of colormap is for data where there are two opposite extremes and a neutral value between those extremes. Examples are vote counts for an election (red/blue Republican/Democrat maps).

To see the colormaps available in matplotlib by default, you can check out this [link](https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html), or the notebook in this repo named Matplotlib_colormaps.

To read more about choosing colormaps, some great references are Tufte's "The Visual Display of Quantitative Information" and [matplotlib's guidence on choosing colormaps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)

```python
figsize=(16,8)
fig, ax = plt.subplots(figsize=figsize)
base = countries.plot(figsize=figsize, color='white', edgecolor='black', linewidth=0.5, ax=ax)
plot_col = 'CONTINENT'
countries.plot(column=plot_col, ax=base, legend=True, cmap='Set2', 
               legend_kwds={'bbox_to_anchor':(0.64, 0.2, 0.5, 0.5), 'title':plot_col})
_ = ax.axis('off')
plt.tight_layout()
_ = ax.annotate('Source: Natural Earth Data (Admin 0 - Countries)',
                xy=(0.045, 0.07), xycoords='figure fraction',
                horizontalalignment='left', verticalalignment='top',
                fontsize=10, color='#555555')
```

### Mapping Sequential Data Features

For sequential-type data features, we'll need a different type of legend as sequential-type data could have a different value for every area on the map. As the main feature of sequential-type data is that it ranges from a minimum positive value to a maximum positive value, a smooth colormap from the minimum value to maximum value can be used to communicate this information, and we can use a colorbar to allow the viewer to determine (roughly) the quantitative value.

```python
figsize=(16,8)
my_cmap='YlGn'
scale=0.5

fig, ax = plt.subplots(figsize=figsize)
base = countries.plot(figsize=figsize, color='white', edgecolor='black', linewidth=0.5, ax=ax)
plot_col = 'POP_EST'
countries.plot(column=plot_col, ax=base, cmap=my_cmap, 
               legend_kwds={'title':plot_col})
_ = ax.axis('off')
vmin = countries[plot_col].min()
vmax = countries[plot_col].max()
sm = plt.cm.ScalarMappable(cmap=my_cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cax = fig.add_axes([1, 0.06, 0.03, 0.82])
cbar = fig.colorbar(sm, shrink=scale, cax=cax)
cbar.ax.set_title('Population', pad=20)
plt.tight_layout()
_ = ax.annotate('Source: Natural Earth Data (Admin 0 - Countries)',
                xy=(0.05, 0.05), xycoords='figure fraction',
                horizontalalignment='left', verticalalignment='top',
                fontsize=10, color='#555555')
```

The extremely large populations of China and India make it hard to make out the populations in other countries; they all just look tiny. Let's adjust the scale from linear to logarithmic.

```python
figsize=(16,8)
my_cmap='YlGn'
scale=0.5
warnings.filterwarnings("ignore")

fig, ax = plt.subplots(figsize=figsize)
base = countries.plot(figsize=figsize, color='white', edgecolor='black', linewidth=0.5, ax=ax)
plot_col = 'POP_EST'
vmin = max([countries[plot_col].min(), 1]) # to avoid division by 0 errors
vmax = countries[plot_col].max()
countries.plot(column=plot_col, ax=base, cmap=my_cmap, 
               legend_kwds={'title':plot_col}, norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
_ = ax.axis('off')
sm = plt.cm.ScalarMappable(cmap=my_cmap, norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
sm._A = []
cax = fig.add_axes([1, 0.06, 0.03, 0.82])
cbar = fig.colorbar(sm, shrink=scale, cax=cax)
cbar.ax.set_title('Population', pad=8)
plt.tight_layout()
_ = ax.annotate('Source: Natural Earth Data (Admin 0 - Countries)',
                xy=(0.05, 0.05), xycoords='figure fraction',
                horizontalalignment='left', verticalalignment='top',
                fontsize=10, color='#555555')
```

# Aggregating Areas: New Boundaries

If you want to create new boundaries from the existing boundaries, you can do that with the dissolve() function. Here, we'll produce a map just of the continents.

```python
figsize=(16,8)
fig, ax = plt.subplots(figsize=figsize)
base = countries.dissolve(by='CONTINENT').plot(figsize=figsize, color='white', edgecolor='black', linewidth=0.5, ax=ax)
_ = ax.axis('off')
plt.tight_layout()
_ = ax.annotate('Source: Natural Earth Data (Admin 0 - Countries)',
                xy=(0.05, 0.05), xycoords='figure fraction',
                horizontalalignment='left', verticalalignment='top',
                fontsize=10, color='#555555')
```

# Aggregating Areas

If you want to aggregate the data within boundaries, you can also do that with the dissolve() function. Here, we'll map aggregate GDP by continent.

```python
figsize=(16,8)
my_cmap='YlGn'
scale=0.5
plot_col = 'GDP_MD_EST'
warnings.filterwarnings("ignore")
geo_df = countries[['CONTINENT', 'geometry', plot_col]].dissolve(by='CONTINENT', aggfunc='sum')

fig, ax = plt.subplots(figsize=figsize)
base = geo_df.plot(figsize=figsize, color='white', edgecolor='black', linewidth=0.5, ax=ax)


# to avoid division by 0 errors and prevent countries with a population smaller than Greenland's from washing out the scale
vmin = max([countries[plot_col].min(), 5*10**5]) 
vmax = countries[plot_col].max()

geo_df.plot(column=plot_col, ax=base, cmap=my_cmap, norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
_ = ax.axis('off')
sm = plt.cm.ScalarMappable(cmap=my_cmap, norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
sm._A = []
cax = fig.add_axes([1, 0.06, 0.03, 0.82])
cbar = fig.colorbar(sm, shrink=scale, cax=cax)
cbar.ax.set_title('Combined GDP\n(in \$millions)', pad=8)
plt.tight_layout()
_ = ax.annotate('Source: Natural Earth Data (Admin 0 - Countries)',
                xy=(0.05, 0.05), xycoords='figure fraction',
                horizontalalignment='left', verticalalignment='top',
                fontsize=10, color='#555555')
```

```python
figsize=(16,8)
my_cmap='YlGn'
scale=0.5
plot_col = 'GDP_MD_EST'
warnings.filterwarnings("ignore")
geo_df = countries

fig, ax = plt.subplots(figsize=figsize)
base = geo_df.plot(figsize=figsize, color='white', edgecolor='black', linewidth=0.5, ax=ax)


# to avoid division by 0 errors and prevent countries with a population smaller than Greenland's from washing out the scale
vmin = max([countries[plot_col].min(), 5*10**3]) 
vmax = countries[plot_col].max()

geo_df.plot(column=plot_col, ax=base, cmap=my_cmap, norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
_ = ax.axis('off')
sm = plt.cm.ScalarMappable(cmap=my_cmap, norm=mpl.colors.LogNorm(vmin=vmin, vmax=vmax))
sm._A = []
cax = fig.add_axes([1, 0.06, 0.03, 0.82])
cbar = fig.colorbar(sm, shrink=scale, cax=cax)
cbar.ax.set_title('GDP (in \$millions)', pad=8)
plt.tight_layout()
_ = ax.annotate('Source: Natural Earth Data (Admin 0 - Countries)',
                xy=(0.05, 0.05), xycoords='figure fraction',
                horizontalalignment='left', verticalalignment='top',
                fontsize=10, color='#555555')
```
